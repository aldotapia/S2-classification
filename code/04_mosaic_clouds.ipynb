```python
import os
import rioxarray as rxr
from rasterio.merge import copy_max
from rioxarray.merge import merge_arrays
```

```python
# Open the data
root_path = '/Volumes/TOSHIBA EXT/Msc/S2_Clouds'
# list files in root_path with .tif ending
files = [os.path.join(root_path, f) for f in os.listdir(root_path) if f.endswith('.tif')]
```

```python
# remove the files with ._ in the name
files = [f for f in files if '._' not in f]
```

```python
sensing_dates = [val.split('/')[-1][:8] for val in files]
```

```python
# create a list of unique dates
unique_dates = list(set(sensing_dates))
```

```python
# filter the files for each date
for date in unique_dates:
    # filter the files for each date
    files_date = [val for val in files if date in val.split('/')[-1][:8]]

    # open the files
    ds = [rxr.open_rasterio(f) for f in files_date]

    # merge the files
    ds_merged = merge_arrays(ds, method='max')

    # save the merged file
    ds_merged.rio.to_raster(f'/Volumes/TOSHIBA EXT/Msc/S2_Clouds_Merged/{date}.tif')

    # close the files
    ds_merged.close()
    for d in ds:
        d.close()

    print(f'{date} done')
```
