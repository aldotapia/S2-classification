```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import matplotlib.patches as patches
```

```python
df = pd.read_csv('../assets/products_query.csv')
```

```python
df['tile'] = [val.split('_')[5] for val in df['title']]
```

```python
df
```

```python
# convert to datetime
df['beginposition'] = pd.to_datetime(df['beginposition'])
```

```python
# create a line plot with X= beginposition, Y=cloudcoverpercentage, color=granule
plt.figure(figsize=(10, 5))
plt.rcParams['figure.dpi'] = 300
plt.rcParams['savefig.dpi'] = 300
sns.set_style("darkgrid")

sns.lineplot(x='beginposition', y='cloudcoverpercentage', hue='tile', data=df)
plt.ylabel('Cobertura de nubes (%)')
plt.xlabel('Fecha')
#plt.savefig("../document/Thesis/figures/clouds.pdf", bbox_inches = 'tight')
plt.show()
```
